```python
import os

path = '.'
dir_list = os.listdir(path)

filt_list = []

for i in dir_list:
    if i.startswith('oasis_cross-sectional'):
        filt_list.append(path + '//' + i)
```

```python
filt_list
```

```python
import tarfile

for i in filt_list:
    # open file
    file = tarfile.open(i)

    # extracting file
    file.extractall('.//Dataset')

    file.close()
    print('File Unzip Complete ', i) 
```
